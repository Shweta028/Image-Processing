```python
import numpy as  np
import matplotlib.pyplot as plt
import cv2

img1 = cv2.imread('lion1.jpg', 1)
img2 = cv2.imread('cat11.jpg', 1)

img1 = cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
img2 = cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)

print(img1.shape)
print(img2.shape)

add = img1 + 50

sub = img1 - 50

titles = ['Leopard', "6 Cats", 'ADDITION', 'SUBTRACTION']
images = [img1, img2, add, sub]

print()

for i in range(4):
    plt.subplot(1, 4, i+1)
    plt.imshow(images[i])
    plt.title(titles[i])
    plt.xticks([])
    plt.yticks([])
    
plt.show()

```
