```python
#Importing opencv and numpy
import cv2
import numpy as np

#Reading the image 
img=cv2.imread('lion.jpg')

#converting the image from BGR(Blue Green Red) to HSV(Hue Saturation Value)
#It takes the image as arguments and COLOR_BGR2HSV attribute 
hsv=cv2.cvtColor(img,cv2.COLOR_BGR2HSV)

#Displaying the original image
cv2.namedWindow('Original Image',cv2.WINDOW_NORMAL)
cv2.imshow('Original Image',img)
cv2.waitKey(1000)           #window hold time in milliseconds

#Now we want to extract only one color from the given image
#For this we have to give minimum and maximum range of particular color
#Let us first extract red color

min_red=np.array([0,100,0])     #Setting minimum and maximum range of red color in HSV
max_red=np.array([5,255,255])

#To detect red color we have to find mask for red color from the hsv image
mask_red=cv2.inRange(hsv,min_red,max_red)
 

# 1. To extract(display) only red color and discard the rest of the image, we perform masking of image i.e bitwise_AND of image and the mask
red_img=cv2.bitwise_and(img,img,mask=mask_red)

#Displaying red color image
cv2.namedWindow('Red Color Extracted',cv2.WINDOW_NORMAL)
cv2.imshow("Red Color Extracted",red_img)
cv2.waitKey(1000)


# 2. To extract only green color and discard the rest of the image, we find mask for green color and perform bitwise_AND
min_green=np.array([50,100,0])
max_green=np.array([80,255,255])

mask_green=cv2.inRange(hsv,min_green,max_green)
green_img=cv2.bitwise_and(img,img,mask=mask_green)

#Displaying green color image
cv2.namedWindow('Green Color Extracted',cv2.WINDOW_NORMAL)
cv2.imshow("Green Color Extracted",green_img)
cv2.waitKey(1000)

# 3. Extracting blue color from given image
min_blue=np.array([78,158,124])
max_blue=np.array([138,255,255])
mask_blue=cv2.inRange(hsv,min_blue,max_blue)
blue_img=cv2.bitwise_and(img,img,mask=mask_blue)

#Displaying blue color image
cv2.namedWindow('Blue Color Extracted',cv2.WINDOW_NORMAL)
cv2.imshow("Blue Color Extracted",blue_img)
cv2.waitKey(1000)

# 4. Extracting yellow color from given image
min_yellow=np.array([20,100,100])
max_yellow=np.array([30,255,255])

mask_yellow=cv2.inRange(hsv,min_yellow,max_yellow)
yellow_img=cv2.bitwise_and(img,img,mask=mask_yellow)

#Displaying yellow color image
cv2.namedWindow('yellow Color Extracted',cv2.WINDOW_NORMAL)
cv2.imshow("yellow Color Extracted",yellow_img)
cv2.waitKey(3000)

cv2.destroyAllWindows()

```
